### Ques1. To get placed with highest salary which degree should be Opted?

## Label Encoding the columns

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dff['status']=le.fit_transform(dff['status'])
dff['hsc_b']=le.fit_transform(dff['hsc_b'])
dff['workex']=le.fit_transform(dff['workex'])
dff['gender']=le.fit_transform(dff['gender'])
dff['ssc_b']=le.fit_transform(dff['ssc_b'])
dff['hsc_s']=le.fit_transform(dff['hsc_s'])
dff['degree_t']=le.fit_transform(dff['degree_t'])
dff['specialisation']=le.fit_transform(dff['specialisation'])
```

## Assigning Independent and Dependent Variables


```python
X = dff[['gender', 'hsc_b', 'hsc_s', 'degree_p', 'workex', 'etest_p',
       'specialisation']]
y = dff[['status']]
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

### 1. SVM

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred1 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1, y_test)
```

### 2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
y_pred2 = logreg.predict(X_test)
```

```python
logreg.score(X_test, y_test)
```

### 3. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KN=KNeighborsClassifier()
```

```python
KN.fit(X_train,y_train)
```

```python
y_pred3=KN.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

## Comparing Accuracies of 3 models

We see the best model is **Logistic Regression** with **77.77%** accuracy, followed by **KNN Classification** with **75.92%** accuracy and **Support Vector Machine (SVM)** with **68.51%** accuracy.
